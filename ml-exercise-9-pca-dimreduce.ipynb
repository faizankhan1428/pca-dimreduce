## Objective
Understand and apply Principal Component Analysis (PCA) to reduce dimensionality, visualize high-dimensional data, and retain maximum variance.

## Step 1: Imports & Dataset

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize features
X_scaled = StandardScaler().fit_transform(X)
```

## Step 2: Apply PCA

```python
# Reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(f"Explained Variance Ratio: {pca.explained_variance_ratio_}")
```

## Step 3: Visualize Transformed Data

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.colorbar(scatter, label='Digit Label')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Digits Dataset')
plt.tight_layout()
plt.show()
```

## âœ… Summary
PCA helps simplify complex datasets while preserving important patterns.

Reducing to 2D allows easy visualization.

Explained variance shows how much info is retained.
